# Train , Vaild DataSet (before spliting)

```python
import json
import os
import csv


# 데이터 디렉토리 경로 지정
data_path = "/content/drive/MyDrive/KOBERT/Training"

# 파일 리스트에서 파일 이름을 하나씩 가져와 처리
file_list = [filename for filename in os.listdir(data_path) if filename.endswith('.json')]

with open("/content/drive/MyDrive/KOBERT/training.csv", 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['newsID', 'sent_num1', 'sent_num2', 'sentence1', 'sentence2', 'gold_label'])

    for filename in file_list:
        # # filename in file_list[:]
        # 전체 파일 경로 생성
        full_file_path = os.path.join(data_path, filename)

        try:
            # 파일 열기
            with open(full_file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)

                sentence_info = data['labeledDataInfo'].get('processSentenceInfo', [])  # 문장 정보 리스트 가져오기
                sentence_contents = [sentence['sentenceContent'].replace("\\", "")  for sentence in data['labeledDataInfo']['processSentenceInfo']]
                #replace로 \표시 없애주기
                consistencies = [sentence['subjectConsistencyYn'] for sentence in data['labeledDataInfo']['processSentenceInfo']]
                # 일관성 N0 인 첫번째 문장 찾기
                consist_no_index = []
                for idx in range(len(sentence_contents)):
                    if consistencies[idx] == 'N':
                        consist_no_index.append(idx)
                noidx = consist_no_index[0]

                # 문장 전체 비교 (순서대로)
                for idx in range(len(sentence_contents) - 1):  # 마지막 문장 제외

                    newsID = data['sourceDataInfo']['newsID'].split('_')[-1]

                    sent_num1 = sentence_info[idx]['sentenceNo']
                    sent_num2 = sentence_info[idx + 1]['sentenceNo']
                    sentence1 = sentence_contents[idx]
                    sentence2 = sentence_contents[idx + 1]
                    consistency1 = consistencies[idx]
                    consistency2 = consistencies[idx + 1]

                    # 두 문장 중 하나라도 'N'이면 'N'으로 설정
                    if consistency1 == 'N' and consistency2 =='N':
                        gold_label = 'Y'
                    elif consistency2 == 'N':
                        gold_label = 'N'
                    else:
                        gold_label = 'Y'

                    row = [newsID, sent_num1, sent_num2, sentence1, sentence2, gold_label]
                    csv_writer.writerow(row)


                # 문장 첫번째로 'N'가 나온것과 그 전전에 있는 문장들 까지 비교후 gold_label 'N' 출력
                for k in range(noidx - 1) : #13번이 no 이면 11번 까지

                    newsID = data['sourceDataInfo']['newsID'].split('_')[-1]

                    sent_num1 = sentence_info[k]['sentenceNo']
                    sent_num2 = sentence_info[noidx]['sentenceNo']
                    sentence1 = sentence_contents[k]
                    sentence2 = sentence_contents[noidx]
                    consistency1 = consistencies[k]
                    consistency2 = consistencies[noidx]

                    if consistency2 == 'N':
                        gold_label = 'N'

                    row = [newsID, sent_num1, sent_num2, sentence1, sentence2, gold_label]
                    csv_writer.writerow(row)


        except FileNotFoundError:
            print(f"파일을 찾을 수 없음: {full_file_path}")
        except json.JSONDecodeError:
            print(f"JSON 디코딩 오류: {full_file_path}")

print('CSV 파일이 생성되었습니다.')
```
